## Ejercicio 03 (a)

```python
import numpy as np
import matplotlib.pyplot as plt

def funcion1(x:float)->float:
    return x
def funcion2(x:float)->float:
    return np.tan(x)

x = np.linspace(-5, 5, 100)
y1 = funcion1(x)
y2 = funcion2(x)
plt.figure(figsize=(5, 4))
plt.plot(x, y1, label='f1(x) = x')
plt.plot(x, y2, label='f2(x) = tan(x)')
ax = plt.gca()
ax.set_ylim([-7, 7])
ax.set_xlim([-6, 6])

plt.title('Gráfica de x y tan(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

## Ejercicio 03 (b)

```python
def funcion3(x:float)->float:
    return x - np.tan(x)

x = np.linspace(-10, 10, 500)
y3 = funcion3(x)
plt.figure(figsize=(5, 4))
plt.plot(x, y3, color='purple', label='f3(x) = x - tan(x)')
ax = plt.gca()
ax.set_ylim([-5, 5])
ax.set_xlim([-3, 3])

plt.title('Gráfica de f3(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np

def bisection_method(f, a, b, tol):
    if f(a) * f(b) > 0:
        raise ValueError("f(a) y f(b) deben tener signos opuestos")

    iterations = 0
    while (b - a) / 2 > tol:
        puntoMedio = (a + b) / 2
        iterations += 1
        if f(puntoMedio) == 0:
            return puntoMedio, iterations
        elif f(a) * f(puntoMedio) < 0:
            b = puntoMedio
        else:
            a = puntoMedio

    return (a + b) / 2, iterations

# Definimos la función
def f(x):
    return x - np.tan(x)

# Definimos los intervalos y la tolerancia
intervalos = [(2,4.5)]
tolerancia = 0.00001

# Calculamos las raíces para cada intervalo
raices = []
for interval in intervalos:
    raiz, iteraciones = bisection_method(f, interval[0], interval[1], tolerancia)
    raices.append(raiz)
    print(f"Raíz encontrada: {raiz}, Iteraciones: {iteraciones}")

#print(raices[0])
```
