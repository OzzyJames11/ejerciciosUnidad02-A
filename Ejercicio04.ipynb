## Ejercicio 4 (a)

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return (x**2 -1)

x = np.linspace(-10, 10, 500)
y = f(x)
plt.figure(figsize=(5, 3))
plt.plot(x, y, label='f(x) = x^2 -1')

plt.title('Gr√°fica de f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
def f(x):
    return np.e**(1-x**2)

x = np.linspace(-10, 10, 500)
y = f(x)
plt.figure(figsize=(5, 3))
plt.plot(x, y, label='f(x) = e^(1-x^2)')
plt.title('Gr√°fica de la funci√≥n f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

## Ejercicio 4 (b)

```python
def bisection_method(f, a, b, tol):
    if f(a) * f(b) > 0:
        raise ValueError("f(a) y f(b) deben tener signos opuestos")

    while (b - a) / 2 > tol:
        puntoMedio = (a + b) / 2
        if f(puntoMedio) == 0:
            return puntoMedio
        elif f(a) * f(puntoMedio) < 0:
            b = puntoMedio
        else:
            a = puntoMedio

    return (a + b) / 2

# Definimos la funci√≥n
def f(x):
    return x**2 - 1 - np.e**(1 - x**2)


# Se definen los par√°metros para encontrar la aproximaci√≥n
bisection_method(f,-2,0,0.001)
```
