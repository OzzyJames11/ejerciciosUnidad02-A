### Se grafica para ver el comportamiento de la gráfica.

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return (x + 3) * (x + 1)**2 * x * (x - 1)**3 * (x - 3)

x = np.linspace(-1.6, 2.6, 500)
y = f(x)
plt.figure(figsize=(5, 3))
plt.plot(x, y, label='f(x) = sin(x)')

plt.title('Gráfica de f(x) = (x + 3) * (x + 1)**2 * x * (x - 1)**3 * (x - 3)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

### Se aplica el método de la bisección

```python
def bisection_method(f, a, b, tol):
    if f(a) * f(b) > 0:
        raise ValueError("f(a) y f(b) deben tener signos opuestos")

    while (b - a) / 2 > tol:
        puntoMedio = (a + b) / 2
        if f(puntoMedio) == 0:
            return puntoMedio
        elif f(a) * f(puntoMedio) < 0:
            b = puntoMedio
        else:
            a = puntoMedio

    return (a + b) / 2

# Definimos la función
def f(x):
    return (x + 3) * (x + 1)**2 * x * (x - 1)**3 * (x - 3)

# Se definen todos los intervalos
intervalos = [(-1.5, 2.5),(-0.5, 2.4),(-0.5, 3),(-3, -0.5)]
tolerancia = 0.00001

# Se calcula todas las raíces para los intervalos
# raices = []
# for intervalo in intervalos:
#   raiz = bisection_method(f, intervalo[0], intervalo[1], tolerancia)
#   raices.append(raiz)

# print(raices)


# Se definen los parámetros para encontrar la aproximación
print(bisection_method(f,-0.5,3,tolerancia))
print(bisection_method(f,-3,-0.5,tolerancia))
```
